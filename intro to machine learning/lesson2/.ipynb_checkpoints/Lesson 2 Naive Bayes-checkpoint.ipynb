# 17. Getting Started with Sci-kit learn

**1.9 Naive Bayes**

Fomular: <http://scikit-learn.org/stable/modules/naive_bayes.html>

Code: <http://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html>

Naive Bayse Classifier

```python
def classify(features_train, labels_train):   
    ### import the sklearn module for GaussianNB
    ### create classifier
    ### fit the classifier on the training features and labels
    ### return the fit classifier
    
    
    ### your code goes here!
    
    from sklearn.naive_bayes import GaussianNB
    clf = GaussianNB()
    clf.fit(features_train, labels_train)
    return clf
```
